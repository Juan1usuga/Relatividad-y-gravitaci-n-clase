```python
import einsteinpy.symbolic as es
import sympy as sp
import numpy as np
from IPython.display import display
```

```python
T,x,y,z = sp.symbols('T x y z')
gmat= np.diag([1,-1,-1,-1   ])
g= es.MetricTensor(gmat, [T,x,y,z])
display(g.tensor())
```

```python
Gama= es.ChristoffelSymbols.from_metric(g)
display(Gama.tensor())
```

```python
T,r,teta,z = sp.symbols('T r teta z')
gmat= np.diag([1,-1,-r**2,-1   ])
g= es.MetricTensor(gmat, [T,r,teta,z])
Gama= es.ChristoffelSymbols.from_metric(g)
display(Gama.tensor())
```

```python
Riemann= es.RiemannCurvatureTensor.from_metric(g)
display(Riemann.tensor())
```

Metrica con curvatura

```python
#Metrica de newton en coordenadas esfericas (r,teta,fi)
#elemento de linea espacial ds^2=dr^2+r^2*dtheta^2+r^2*sin^2(theta)*dphi^2
T, r, teta, fi = sp.symbols('T r teta fi')
G, M, c = sp.symbols('G M c')
gmat =np.diag([1-2*G*M/r/c**2, -1, -r**2, -r**2*sp.sin(teta)**2])
g= es.MetricTensor(gmat, [T,r,teta,fi])
display(g.tensor())
```

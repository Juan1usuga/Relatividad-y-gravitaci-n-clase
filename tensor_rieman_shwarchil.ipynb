```python
import einsteinpy.symbolic as es
import sympy as sp
import numpy as np
from IPython.display import display
```

```python
T,x,y,z = sp.symbols('T x y z')
gmat= np.diag([1,-1,-1,-1   ])
g= es.MetricTensor(gmat, [T,x,y,z])
display(g.tensor())
```

```python
Gama= es.ChristoffelSymbols.from_metric(g)
display(Gama.tensor())
```

```python
T,r,teta,z = sp.symbols('T r teta z')
gmat= np.diag([1,-1,-r**2,-1   ])
g= es.MetricTensor(gmat, [T,r,teta,z])
Gama= es.ChristoffelSymbols.from_metric(g)
display(Gama.tensor())
```

```python
Riemann= es.RiemannCurvatureTensor.from_metric(g)
display(Riemann.tensor())
```

Metrica con curvatura

```python
#Metrica de newton en coordenadas esfericas (r,teta,fi)
#elemento de linea espacial ds^2=dr^2+r^2*dtheta^2+r^2*sin^2(theta)*dphi^2
T,r,teta,fi, M,G,c = sp.symbols('T r theta phi G M c')
gmat =np.diag([1-2*G*M/r/c**2, -1, -r**2, -r**2*sp.sin(teta)**2])
g= es.MetricTensor(gmat, [T,r,teta,fi])
display(g.tensor())
```

```python
Riemann= es.RiemannCurvatureTensor.from_metric(g)
Rtensor=Riemann.tensor()
display(Rtensor)
```

```python
# los indices del tensor de Riemann se ordenan de la siguiente manera: los primeros 2 son los internos y los otros 2 son los externos de la matriz de matrices 
display(Rtensor[1,0,0,1]) , display(Rtensor[0,1,1,0])
display(Rtensor[0,1,0,1]) , display(Rtensor[1,0,1,0])
display(Rtensor[1,0,0,1]) , display(Rtensor[0,1,1,0]) 
```

 obtener el tensor completamente covariante

```python
gRiemann_gonorrea = sp.tensorproduct(g.tensor(),Riemann.tensor())
Riemann_cuca =sp.tensorcontraction(gRiemann_gonorrea, (1,2))
display(Riemann_cuca)
```

```python
Riemann_cov =Riemann.change_config('llll')
Rcov= Riemann_cov.tensor()
display(Rcov)
```

```python
display(Rcov[1,0,0,1]) , display(Rcov[0,1,1,0])
```

Tensor de Ricci

```python
# con sympy
Riemann_cov.tensor()
```

```python
ginv = sp.Matrix(g.tensor()).inv()
ginv
```

```python
ginv = g.change_config('uu')
display(ginv.tensor())
```

```python
gRiemann_ricci= sp.tensorproduct(g.tensor(),Riemann_cov.tensor())
Ricci_tensor=sp.simplify(sp.tensorcontraction(gRiemann_ricci, (0,2),(1,5)))
display(Ricci_tensor)
```

```python
# con Einstienpy
Ricci = es.RicciTensor.from_metric(g)
display(Ricci.tensor())
```

```python
# la traza del tensor de Ricci 
gRicci = sp.tensorproduct(ginv.tensor(),Ricci.tensor())
R= sp.simplify(sp.tensorcontraction(gRicci, (0,2),(1,3)))
R # curvatura de Riemann 
```

```python
import einsteinpy.symbolic as es
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

```

```python
g,x,c = sp.symbols('g x c')

gmat= np.diag([1+2*g*x/c**2, -1, -1, -1])
gmat
```

```python
coordenadas = sp.symbols('T x y z') # T= tc 
gmet = es.MetricTensor(gmat, coordenadas)
gmet
```

```python
gtensor = gmet.tensor()
gtensor
```

```python
G = es.ChristoffelSymbols.from_metric(gmet)
G.tensor()
```

```python
G[0]
```

```python
G[0,0,1] 
```

```python
xpp=sp.symbols("\ddot{T} \ddot{x} \ddot{y} \ddot{z}")
xp= sp.symbols("\dot{T} \dot{x} \dot{y} \dot{z}")
```

```python
for sigma in range(4):
    ecuacion = 0
    for i in range(4):
        for mu in range(4):
            for nu in range(4):
                ecuacion += -G[i,mu,nu]*xp[mu]*xp[nu]
            
    ecuacion= sp.Eq(xpp[sigma], ecuacion)
    print(ecuacion)
```

```python
def edg_galileo(tau, Ys, g, c):
    T, x, y, z, Ut, Ux, Uy, Uz = Ys
    #derivadas 
    dt_dtau = Ut
    dx_dtau = Ux
    dy_dtau = Uy
    dz_dtau = Uz
    
    #derivadas de U
    dUt_dtau = -2*Ut*Ux*g/(c**2 +2*g*x)
    dUx_dtau = -Ut**2*g/c**2
    dUy_dtau = 0
    dUz_dtau = 0 
    return [dt_dtau, dx_dtau, dy_dtau, dz_dtau, dUt_dtau, dUx_dtau, dUy_dtau, dUz_dtau]
```

```python
edg_galileo(1, [1,1,1,1,1,1,1,1], 9.8, 1)
```
